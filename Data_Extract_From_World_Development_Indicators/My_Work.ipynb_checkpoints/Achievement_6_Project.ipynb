## Step 1: Import Necessary Libraries

```python
# Import Standard Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from os.path import join
```

## Step 2: Set Path and Load Data

```python
# Set Path

path = r'../Data_Extract_From_World_Development_Indicators'
```

```python
# Load Data 

df_world_bank = pd.read_csv("../Data/WDI_Data.csv")
```

## Step 3: Initial Inspection and Understanding 

```python
# Check the dimensions of data

df_world_bank.shape

print("/nColumn info:")
df_world_bank.info()

print("/nBasic statistic:")
df_world_bank.describe()
```

```python
# Check Data

print(f"Initial shape: {df_world_bank.shape}")
df_world_bank.head()
```

## Step 4: Data Cleaning

```python
# Inspect Columns and Missing Data

print("Columns:/n", df_world_bank.columns.tolist())
print("/nMissing values per column:/n", df_world_bank.isna().sum())
```

#### Handle Missing values

```python
# Drop Missing Values

df_world_bank = df_world_bank.dropna(subset=[ "Country Name", "Country Code", "Series Name", "Series Code"])
print(f"Shape after dropping core missing values rows: {df_world_bank.shape}")
```

#### Reshape Long Format

```python
# Identify all year columns

year_cols = [col for col in df_world_bank.columns if "[YR" in col]

id_vars = ["Country Name", "Country Code", "Series Name", "Series Code"]

df_long = pd.melt(df_world_bank, id_vars=id_vars, var_name= "Year", value_name= "Value")
```

```python
# Check for results

df_long.columns
```

```python
df_world_bank.head
```

#### Convert Year Clomuns to Strings

```python
# Convert columns

df_long['Year'] = df_long['Year'].astype(str).str.extract(r'(\d{4})').astype(int)
```

```python
# Cross Check

df_long.head(25)
```

```python
# Cross Check

df_long.info()
```

## Step 5: Deep Dive - Understanding the Data

```python
#Check for Indicators

indicators = [
    "GDP per capita (current US$)",
    "Life expectancy at birth, total (years)",
    "Individuals using the Internet (% of population)"
]

df_filtered = df_long[df_long['Series Name'].isin(indicators)]

# Check the filtered data
df_filtered.head()
```

#### Back to Wide for Analysis

```python
# Convert Value to numeric, coerce errors to NaN
df_filtered['Value'] = pd.to_numeric(df_filtered['Value'], errors='coerce')

# Check the column type
print(df_filtered['Value'].dtype)
print(df_filtered['Value'].isna().sum())
```

```python
# Pivot the table to have indictors as columns

df_wide = df_filtered.pivot_table(
    index=['Country Name', 'Country Code', 'Year'],
    columns='Series Name',
    values='Value'
).reset_index()

# Optional: remove column grouping from pivot
df_wide.columns.name = None

df_wide.head(25)
```

```python
# Cross Check
df_wide.columns
```

```python
# Examples 

df_wide[df_wide['Country Name'] == 'Argentina']
```

```python
# Group GPD per capital 

df_wide.groupby('Year')['GDP per capita (current US$)'].mean()
```

```python
# Visual Exploration

arg = df_wide[df_wide['Country Name'] == 'Argentina']
plt.plot(arg['Year'], arg['Life expectancy at birth, total (years)'])
plt.title("Life Expectancy in Argentina")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.savefig("Life Expectancy in Argentina.png")
plt.show()
```

```python
# Example: Compare multiple countries
countries = ['Argentina', 'Brazil', 'Chile',]
df_plot = df_wide[df_wide['Country Name'].isin(countries)]
for c in countries:
    plt.plot(df_plot[df_plot['Country Name'] == c]['Year'],
             df_plot[df_plot['Country Name'] == c]['GDP per capita (current US$)'],
             label=c)
plt.legend()
plt.title("GDP per capita Comparison")
plt.xlabel("Year")
plt.ylabel("GDP per capita (US$)")
plt.savefig("GDP per capita Comparison.png")
plt.show()

```

```python
# GDP per capita growth
df_wide['GDP_growth'] = df_wide.groupby('Country Name')['GDP per capita (current US$)'].pct_change() * 100

# Life Expectancy change
df_wide['LE_change'] = df_wide.groupby('Country Name')['Life expectancy at birth, total (years)'].diff()

# Internet usage growth
df_wide['Internet_growth'] = df_wide.groupby('Country Name')['Individuals using the Internet (% of population)'].pct_change() * 100
```

```python
# Example for Argentina
country = 'Argentina'
df_country = df_wide[df_wide['Country Name'] == country][
    ['GDP_growth', 'LE_change', 'Internet_growth']
]

# Compute correlation matrix
corr_matrix = df_country.corr()
print(f"Correlation matrix for {country}:\n", corr_matrix)
```

```python
# Filter for a single country
country = 'Argentina'
df_country = df_wide[df_wide['Country Name'] == country]

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# GDP growth (left axis)
ax1.plot(df_country['Year'], df_country['GDP_growth'], color='blue', marker='o', label='GDP Growth (%)')
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP Growth (%)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Life Expectancy change (right axis)
ax2 = ax1.twinx()
ax2.plot(df_country['Year'], df_country['LE_change'], color='green', marker='x', label='Life Expectancy Change')
ax2.set_ylabel('Life Expectancy Change (years)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Internet usage growth (left axis, dashed line)
ax1.plot(df_country['Year'], df_country['Internet_growth'], color='red', linestyle='--', marker='s', label='Internet Growth (%)')

# Combine legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper left')

plt.title(f"Trends for {country}: GDP Growth, Life Expectancy Change, Internet Usage Growth")
plt.show()
plt.savefig("Trends for Argentina.png")
```

```python
sns.pairplot(df_country, kind='reg')
plt.suptitle(f"Scatter Plots of Indicators for {country}", y=1.02)
plt.savefig("Indicators for Argentina.png")
plt.show()
```

# Step 6: Analyze Specific Questions

### Question 1: How has GDP per capita changed over time in selected countries?

```python
# Answer 1: Code a& Plot

countries = ['Argentina', 'Brazil', 'Chile']

plt.figure(figsize=(10,6))
for country in countries:
    df_country = df_wide[df_wide['Country Name'] == country]
    plt.plot(df_country['Year'], df_country['GDP_growth'], marker='o', label=country)

plt.title('GDP per Capita Growth (%)')
plt.xlabel('Year')
plt.ylabel('GDP Growth (%)')
plt.legend()
plt.grid(True)
plt.savefig("GDP per capita Growth.png")
plt.show()
```

##### Interpretation: GDP growth fluctuates; Chile is steadier, Argentina and Brazil more volatile.

### Question 2: How has Life Expectancy evolved and does it correlate with economic growth?

```python
# Answer 2: Code and Plot

plt.figure(figsize=(10,6))
for country in countries:
    df_country = df_wide[df_wide['Country Name'] == country]
    plt.plot(df_country['Year'], df_country['LE_change'], marker='x', label=country)

plt.title('Life Expectancy Change (years)')
plt.xlabel('Year')
plt.ylabel('Change in Life Expectancy')
plt.legend()
plt.grid(True)
plt.savefig("Life Expectancy Change(years).png")
plt.show()

# Correlation with GDP
for country in countries:
    df_country = df_wide[df_wide['Country Name'] == country][['GDP_growth','LE_change']]
    corr = df_country.corr().iloc[0,1]
    print(f'{country} - Correlation GDP vs Life Expectancy Change: {corr:.2f}')
```

##### Interpretation: Life Expectancy rises gradually; moderate correlation with GDP growth.

### Question 3: How has Internet usage grown across countries and globally?

```python
# Answer 3: Code and Plot

plt.figure(figsize=(10,6))
for country in countries:
    df_country = df_wide[df_wide['Country Name'] == country]
    plt.plot(df_country['Year'], df_country['Internet_growth'], marker='s', linestyle='--', label=country)

plt.title('Internet Usage Growth (%)')
plt.xlabel('Year')
plt.ylabel('Internet Growth (%)')
plt.legend()
plt.grid(True)
plt.savefig("Internet Usage Growth")
plt.show()

# Global average
df_global = df_wide.groupby('Year')[['Internet_growth']].mean()
plt.figure(figsize=(10,5))
plt.plot(df_global.index, df_global['Internet_growth'], marker='o', color='black')
plt.title('Global Average Internet Usage Growth (%)')
plt.xlabel('Year')
plt.ylabel('Internet Growth (%)')
plt.grid(True)
plt.savefig("Global Average Internet Usage Growth(%).png")
plt.show()
```

##### Interpretation: Internet adoption rises sharply, especially after 2010; global trend is consistent.

### Question 4: Are GDP growth, Life Expectancy change, and Internet adoption correlated?

```python
# Answer 4: Code and Plot - Example for Germany

df_country = df_wide[df_wide['Country Name'] == 'Germany'][['GDP_growth','LE_change','Internet_growth']]

# Correlation matrix
corr_matrix = df_country.corr()
print("Correlation matrix for Germany:\n", corr_matrix)

# Scatter plots
sns.pairplot(df_country, kind='reg')
plt.suptitle("Scatter Plots of Indicators for Germany", y=1.02)
plt.savefig("Scatter Plots of Indicators for Germany.png")
plt.show()
```

### Question 5: How do countries compare in terms of development and technology adoption trends?

```python
# Answer 5: Code and Plot - Multi-country comparison

fig, axes = plt.subplots(3, 1, figsize=(12, 15), sharex=True)

# GDP growth
for country in countries:
    df_country = df_wide[df_wide['Country Name'] == country]
    axes[0].plot(df_country['Year'], df_country['GDP_growth'], marker='o', label=country)
axes[0].set_title('GDP per Capita Growth (%)')
axes[0].set_ylabel('GDP Growth (%)')
axes[0].legend()
axes[0].grid(True)

# Life Expectancy
for country in countries:
    df_country = df_wide[df_wide['Country Name'] == country]
    axes[1].plot(df_country['Year'], df_country['LE_change'], marker='x', label=country)
axes[1].set_title('Life Expectancy Change (years)')
axes[1].set_ylabel('LE Change')
axes[1].legend()
axes[1].grid(True)

# Internet usage
for country in countries:
    df_country = df_wide[df_wide['Country Name'] == country]
    axes[2].plot(df_country['Year'], df_country['Internet_growth'], marker='s', linestyle='--', label=country)
axes[2].set_title('Internet Usage Growth (%)')
axes[2].set_xlabel('Year')
axes[2].set_ylabel('Internet Growth (%)')
axes[2].legend()
axes[2].grid(True)

plt.tight_layout()
plt.savefig("GDP,Life Expectancy and internet Usage Globally.png")
plt.show()
```

##### Interpretation: Chile is more stable in GDP and health, while Argentina and Brazil show larger fluctuations. Internet adoption grows rapidly in all countries.
